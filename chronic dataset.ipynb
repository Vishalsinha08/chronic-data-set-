```python
#multivariative 
colos=["age","blood_pressure","blood_glucose_random","serum_creatinine","class"]

g=sns.PairGrid(df[colos],hue="class",palette="coolwarm")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot,cmap="Blues_d")
g.map_diag(sns.histplot)
g.add_legend()
plt.title("pair grid for selected columns")
plt.show()
```
